```python
# Bài 1: Nhập vào số nguyên n, in ra màn hình số gấp đôi của n
n = int(input())
n = 2 * n
print(n)
```

```python
# Bài 2: Một khu đất hình chữ nhật có chiều rộng a, chiều dài b. Người ra xây một khu
# vui chơi hình tròn tại vị trí trung tâm khu đất như hình vẽ. Nhập vào a, b. Hãy tính phần
# diện tích còn để trồng cây xung quanh là bao nhiêu. Biết pi = 3.14, lấy kết quả đến 2
# chữ số phần thập phân.
a, b, r = map(int, input().split())
pi = 3.14
if a < b and a > 0 and b > 0 and r > 0:
  phan_trong_cay = (a * b) - (pi * r * r)
  print(f"{phan_trong_cay:.2f}")
else:
  print("Không hợp lệ")
```

```python
# Bài 8: Nhập vào họ tên chủ hộ và chỉ số điện kế tháng trước và chỉ số điện kế tháng này,
# tính tiền điện tháng này cho hộ biết:
# - Từ 0 – 50kWh: 1.984 đồng/kWh
# - Bậc 2: Từ 51 – 100 kWh: 2.050 đồng/kWh
# - Bậc 3: Từ 101 – 200 kWh: 2.380 đồng/kWh
# - Bậc 4: Từ 201 – 300 kWh: 2.998 đồng/kWh
# - Bậc 5: Từ 301 – 400 kWh: 3.350 đồng/kWh
# - Bậc 6: Từ 401 trở đi kWh: 3.460 đồng/kWh
ten_chu_ho = input()
chi_so_thang_truoc = int(input())
chi_so_thang_nay = int(input())
dien_tieu_thu = chi_so_thang_nay - chi_so_thang_truoc
if 0 < dien_tieu_thu <= 50:
  tien_dien = dien_tieu_thu * 1.984
  tien_thue = tien_dien * 0.08
  tong_tien = tien_dien + tien_thue

if 51 <= dien_tieu_thu <= 100:
  tien_dien = dien_tieu_thu * 2.050
  tien_thue = tien_dien * 0.08
  tong_tien = tien_dien + tien_thue

if 101 <= dien_tieu_thu <= 200:
  tien_dien = dien_tieu_thu * 2.380
  tien_thue = tien_dien * 0.08
  tong_tien = tien_dien + tien_thue

if 201 <= dien_tieu_thu <= 300:
  tien_dien = dien_tieu_thu * 2.998
  tien_thue = tien_dien * 0.08
  tong_tien = tien_dien + tien_thue

if 301 <= dien_tieu_thu <= 400:
  tien_dien = dien_tieu_thu * 3.350
  tien_thue = tien_dien * 0.08
  tong_tien = tien_dien + tien_thue

if dien_tieu_thu >= 401:
  tien_dien = dien_tieu_thu * 3.460
  tien_thue = tien_dien * 0.08
  tong_tien = tien_dien + tien_thue
print(ten_chu_ho)
print(f"Tien Phai Tra La {tong_tien:.3f}")


```

```python
# Bài 4. Kiểm tra tuổi đủ điều kiện bầu cử: Viết chương trình nhập năm sinh của một
# người, kiểm tra xem người đó có đủ 18 tuổi để đi bầu cử hay không.
nam_sinh = int(input())
tuoi = 0
for i in range(nam_sinh, 2027):
  tuoi += 1
if tuoi >= 18:
  print("Du Dieu Kien Di Bau Cu")
else:
  print("Chua Du Dieu Kien Di Bau Cu")
```

```python
# Bài 5:So sánh 2 số: Viết chương trình nhập vào hai số, in ra số lớn hơn hoặc thông báo
# nếu hai số bằng nhau
a, b = map(int, input().split())
if a > b:
  print(a)
elif b > a:
  print(b)
elif a == b:
  print("a Bang b")
```

```python
# Bài 2. Kiểm tra số có tận cùng 5 không: Viết chương trình nhập một số nguyên, kiểm tra
# số đó có tận cùng 5 không, nếu đúng in ra “True”, nếu sai in ra “False”
n = int(input())
if n % 10 == 5:
  print(True)
else:
  print(False)

```

```python
#  Bài 3. Kiểm tra số có chia hết cho 3 và 5 không: Viết chương trình nhập một số nguyên,
# kiểm tra số đó có chia hết cho 3 và 5 không, nếu đúng in ra “True”, nếu sai in ra
# “False”.
n = int(input())
if n % 3 == 0 and n % 5 == 0:
  print(True)
else:
  print(False)
```
