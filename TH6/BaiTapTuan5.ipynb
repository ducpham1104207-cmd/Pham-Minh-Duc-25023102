```python
# W5A3 Viết hàm kiểm tra một số có phải là số nguyên tố hay không?
def so_nguyen_to(n):
  so_nguyen_to = True
  if n < 2:
    so_nguyen_to = False
  else:
    for i in range(2, n):
      if n % i == 0:
        so_nguyen_to = False
      else:
        so_nguyen_to = True
    print("Day Co La So Nguyen To")
n = int(input())
so_nguyen_to(n)

```

```python
# W5A4 VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên.
def perf_num(a):
  tong_uoc = 0
  if a < 1:
    print("Day Khong Phai So Hoan Hao")
  else:
    for i in range(1, a):
      if a % i == 0:
        tong_uoc += i
    if a == tong_uoc:
      print("Day La So Hoan Hảo")
    else:
      print("Day Khong Phai So Hoan Hao")
a = int(input())
perf_num(a)
```

```python
# W5A5 Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử
# đầu tiên có giá trị k trong list số nguyên, nếu không có thì trả về -1
def tim_vi_tri(danh_sach_so_nguyen, k):
  for i in range(len(danh_sach_so_nguyen)):
    if danh_sach_so_nguyen[i] == k:
      return i
  return -1
danh_sach_so_nguyen = list(map(int, input().split()))
k = int(input())
tim_vi_tri(danh_sach_so_nguyen, k)
```

```python
# W5A6 Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!
def giai_thua(n):
  giai_thua = 1
  for i in range(1, n + 1):
    giai_thua = giai_thua * i
  return giai_thua
n = int(input())
giai_thua(n)


```

```python
# W5A7 Viết hàm giả lập một máy tính bỏ túi.
# Hàm thực hiện tính toán kết quả của phép toán giữa số \( num1 \) và \( num2 \) với toán tử \( operat
# \)
# Biết các toán tử đơn giản thực hiện phép toán lần lượt là:
# Phép trừ \( - \)
# Phép cộng \( + \)
# Phép chia \( / \)
# Phép nhân \( * \)
# Kết quả làm tròn đến chữ số thập phân thứ hai.
# Đầu vào: Hai số thực và phép toán cần thực hiện (trong số các phép toán \(+, -, *, /\)).
# Đầu ra: Kết quả phép toán
def may_tinh(num1, num2, operator):
  if operator == "+":
    ket_qua = num1 + num2
    print(f"{ket_qua:.2f}")
  elif operator == "-":
    ket_qua = num1 - num2
    print(f"{ket_qua:.2f}")
  elif operator == "*":
    ket_qua = num1 * num2
    print(f"{ket_qua:.2f}")
  elif operator == "/":
    if num2 == 0:
      print("Khong The Thuc Hien Phep Tinh")
    else:
      ket_qua = num1 / num2
      print(f"{ket_qua:.2f}")
num1 = float(input())
num2 = float(input())
operator = input()
may_tinh(num1, num2, operator)
```

```python
# W5A9 Viết hàm trả về tổng các chữ số của số nguyên dương n
def tong_nguyen_duong(n):
  tong = 0
  while n > 0:
    tong = tong + (n % 10)
    n = n // 10
  return tong
n = int(input())
tong_nguyen_duong(n)



```

```python
# W5A11 Viết hàm is_product_divisible_by_10(a, b) trả về True nếu tích a*b chia hết cho 10, ngược lại False
def is_product_divisible_by_10(a, b):
  if (a * b) % 10 ==0:
    return True
  return False
a, b = map(int, input().split())
is_product_divisible_by_10(a, b)
```

```python
# WA15 Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông
def is_right_triangle(a, b, c):
  if a + b > c and a + c > b and c + b > a:
    return True
  return False
a, b, c = map(int, input().split())
is_right_triangle(a, b, c)
```
