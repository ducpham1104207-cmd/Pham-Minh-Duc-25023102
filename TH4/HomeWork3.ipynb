```python
# Bài 8: Nhập vào số nguyên dương a và b, in toàn bộ ước dương chung của a và b
a, b = map(int, input().split())
for i in range(1, a+1):
  if a % i == 0:
    if b % i == 0:
      print(i)

```

```python
# Bài 8: Nhập vào số nguyên dương a và b, in toàn bộ ước dương chung của a và b
a, b = map(int, input().split())
i = 1
while i <= a and i <= b:
  if a % i == 0 and b % i == 0:
    i += 1
    print(i)
```

```python
# Bài 9: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số chính phương hay
# không? Số chính phương là số có căn bậc hai là một số nguyên VD: 1, 4, 9, 16, 25….
a = int(input())
if a > 0:
  can_bac_hai = a ** 0.5
  if can_bac_hai == int(can_bac_hai):
    print("Day La So Chinh Phuong")
  else:
    print("Day Khong La So Chinh Phuong")
```

```python
# Bài 10: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số hoàn hảo hay không.
# Số hoàn hảo là số có tổng các ước (ước nhỏ hơn nó) bằng chính nó, VD: số 6, 28…
a = int(input())
tong = 0
for i in range(1, a):
  if a % i == 0:
    tong = tong + i
    if tong == a:
      print("Day La So Hoan Hao")
else:
  print("Day Khong La So Hoan Hao")
```

```python
# Bài 10: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số hoàn hảo hay không.
a = int(input())
i = 1
tong = 0
while i < a:
  if a % i == 0:
    tong = tong + i
    i = i + 1
  else:
    i += 1
if tong == a:
  print("Day la so hoan hao")
else:
  print("Day khong phai so hoan hao")

```

```python
# Bài 16 Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ
n = int(input())
count_chan = 0
count_le = 0
tong = 0
if n == 0:
  count_chan = 1
else:
  while n > 0:
    so_cuoi_cung = n % 10
    if so_cuoi_cung % 2 == 0:
      count_chan += 1
      n = n // 10
    else:
      count_le += 1
      n = n // 10
print(count_chan)
print(count_le)

```

```python
# Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt quá A
x = int(input())
f1, f2 = 1, 1
result = 1
while f2 <= x:
    result = f2
    f1, f2 = f2, f1 + f2
print(result)
```
