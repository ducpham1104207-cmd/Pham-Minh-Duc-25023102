Bài 8: Giải hệ phương trình tuyến tính:
ax + by = m
cx + dy = n

```python
a, b, m, c, d, n = map(float, input().split())
i = a * d - b * c
if i == 0:
    print("Hệ phương trình vô nghiệm hoặc vô số nghiệm")
else:
    x = (m * d - b * n) / i
    y = (a * n - m * c) / i
    print(f"x = {x}, y = {y}")
```

Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vàomộtđiểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
x0, y0, r, x, y = map(float, input().split())
d = ((x - x0)**2 + (y - y0)**2)**0.5
if d == r:
    print("Điểm A thuộc đường tròn")
elif d < r:
    print("Điểm A nằm bên trong đường tròn")
else:
    print("Điểm A nằm bên ngoài đường tròn")
```
