```python
# W4A4
# .[CountingDigits]
# Viết chương trình nhập vào một số nguyên n và in ra màn hình số chữ số (trừ dấu)
# của số đó. Không dùng kiểu dữ liệu string
n = abs(int(input()))
count = 0
if n == 0:
  count = 1
else:
  while n > 0:
    n = n // 10
    count = count + 1
print(f"So chu so la {count}")
```

```python
# W4A8
# ReverseAddPalindrome]
# Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số
# đối xứng (palindrome). In ra số bước và giá trị palindrome thu được. 
n = int(input())
n_rev = 0
steps = 0
while True:
  n_rev = 0
  temp = n
  while temp > 0:
    n_rev = n_rev * 10 + temp % 10
    temp = temp // 10
  if n == n_rev:
    break
  else:
    n = n + n_rev
    steps = steps + 1
print(steps, n) 
        
```

```python
# W4A9
# [PerfectSquaresDistinctDigits]
# Nhập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số
# đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu
# cách). Nếu không có, in ra (no number).
n = int(input())
perfect_squares = []

for i in range(1, int(n**0.5) + 1):
    square = i * i
    if square > n:
        break
    digits = str(square)
    if len(set(digits)) == len(digits):
        perfect_squares.append(square)

if perfect_squares:
    print(perfect_squares)
else:
    print("no number")
```

```python
# W4A14
# [GCD]
# Cho hai số m và n là hai số nguyên dương nhập từ bàn phím. Viết chương trình
# tìm ước chung lớn nhất của hai số m và n. 
m, n = map(int, input().split())
while m > 0 and n > 0:
    if m > n:
        m = m - n
    else:
        n = n - m
print(m)    
```

```python
# W4A15
# [ChickenAndDog]
# "Vừa gà vừa chó,
# Bó lại cho tròn,
# Ba mươi sáu con,
# Một trăm chân chẵn"
# Từ bài toán dân gian trên, mở rộng thành chương trình nhận đầu vào là tổng số
# con và tổng số chân của gà và chó. Nếu tìm được số phù hợp, in ra số lượng gà và
# số lượng chó. Ngược lại, in ra "invalid"
tong_con_vat, tong_chan = map(int, input().split())
if (tong_chan % 2 == 0) and (tong_con_vat % 2 == 0) and (tong_chan >= 2 * tong_con_vat) and (tong_chan <= 4 * tong_con_vat):
    so_ga = (4 * tong_con_vat - tong_chan) // 2
    so_cho = tong_con_vat - so_ga
    print(so_ga, so_cho)    
else:
    print("invalid")
```
